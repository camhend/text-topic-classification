```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_dir = '/home/henderc7/cs571/202540_data571_final_project_group16/proj_data/task1_topics'
trainX = np.loadtxt(f'{data_dir}/train.sparseX', dtype=int)
trainX = pd.DataFrame(trainX, columns=['N', 'D', 'V'])
devX = np.loadtxt(f'{data_dir}/dev.sparseX', dtype=int)
devX = pd.DataFrame(devX, columns=['N', 'D', 'V'])

trainY = np.loadtxt(f'{data_dir}/train.CT', dtype=int)
trainY = pd.DataFrame(trainY, columns=['C'])
devY = np.loadtxt(f'{data_dir}/dev.CT', dtype=int)
devY = pd.DataFrame(devY, columns=['C'])
```

```python
trainX
```

```python
train_class_counts = trainY['C'].value_counts()
plt.figure(figsize=(8, 5))
plt.bar(train_class_counts.index, train_class_counts.values)
plt.xlabel("Class")
plt.ylabel("Count")
plt.title("Train Class Distribution")
print(train_class_counts.values)
plt.show()
```

```python
dev_class_counts = devY['C'].value_counts()
plt.figure(figsize=(8, 5))
plt.bar(dev_class_counts.index, dev_class_counts.values)
plt.xlabel("Class")
plt.ylabel("Count")
plt.title("Dev Class Distribution")
print(dev_class_counts.values)
plt.show()
```

```python
train_word_counts = trainX.groupby('D')['V']
train_avg_word_counts = train_word_counts.mean()
train_avg_word_counts
plt.figure(figsize=(8, 5))
plt.bar(train_avg_word_counts.index, train_avg_word_counts.values)
plt.xlabel("Word")
plt.ylabel("Count")
plt.title("Train Avg Word Counts Distribution")
plt.show()
```

```python
dev_word_counts = devX.groupby('D')['V']
dev_avg_word_counts = dev_word_counts.mean()
plt.figure(figsize=(8, 5))
plt.bar(dev_avg_word_counts.index, dev_avg_word_counts.values)
plt.xlabel("")
plt.ylabel("Count")
plt.title("dev Avg Word Counts Distribution")
plt.show()
```

```python
idx0 = devX[devX['D'] == 0]
plt.figure(figsize=(8, 5))
plt.hist(idx0['V'], bins=200)
plt.xlabel("0 dim")
plt.ylabel("Count")
plt.title("0 dim counts")
plt.show()
```
